# First Importing libraries

```python
import cv2
import matplotlib.pyplot as pt
```

## loading the image
## imread() is used to load the image.
## cvtColor() is used to convert an image from one color space to another.

```python
img=cv2.imread("C:\\Users\\Rakshith\\Pictures\\mahi.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
pt.figure(figsize=(9,9))
pt.axis("off")
pt.title("original image")
pt.imshow(img)
```

# Coverting image to Grayscale
## Now we have to convert the image to Grayscale image

```python
im_gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
pt.figure(figsize=(9,9))
pt.axis("off")
pt.imshow(im_gry,cmap="gray")
pt.title("grayscale image")
pt.show()
```

# Now inverting this image
## For inverting bitwise_not() function is used

```python
im_invert=cv2.bitwise_not(im_gry)
pt.figure(figsize=(9,9))
pt.axis("off")
pt.imshow(im_invert,cmap="gray")
pt.title("inverted image")
pt.show()
```

# smoothing the inverted blur image
## Here GaussianBlur() function is used

```python
im_smooth=cv2.GaussianBlur(im_invert, (21,21),sigmaX=0, sigmaY=0)
pt.figure(figsize=(9,9))
pt.axis("off")
pt.imshow(im_smooth,cmap="gray")
pt.title("Smoothen Image")
pt.show()
```

# Finally converting to Pencil sketch
## Here divide()fuction is used

```python
im_sketch=cv2.divide(im_gry,255-im_smooth,scale=255)
pt.figure(figsize=(9,9))
pt.axis("off")
pt.imshow(im_sketch,cmap="gray")
pt.title("final sketch Image")
pt.show()
```
